```python
import pandas as pd
import os
import glob

from lrg_omics.metabolomics.common import metadata_from_worklist, metadata_from_filename


class MetabolomicsQc():
    def __init__(self):
        pass


class MetabolomicsQcGui():
    
    def __init__(self, path):
        self._ms_files = [x for x in glob.glob(path + '/*.mzXML')]              # the list of files in the path directory
        self._filenames = [os.path.basename(x)[:-6] for x in glob.glob(path + '/*.mzXML')]
        self._sizes = [os.path.getsize(self._ms_files[k]) for k in range(len(self._ms_files))]
#         self._worklist = pd.read_csv(glob.glob(path + '/*.csv')[0])             # the worklist!!
        self._worklist = pd.DataFrame()
        
        self._standards = [x for x in glob.glob(path + '/*Standard*.mzXML')]    # only standard samples are includded
        self._sa_pool_files = [x for x in glob.glob(path + '/*SA-Pool*.mzXML')] # only SA-pool samples are includded
        self._mh_pool_files = [x for x in glob.glob(path + '/*MH-Pool*.mzXML')] # only MH-pool samples are includded
        self._qc_files = [x for x in glob.glob(path + '/*QC0*.mzXML')]          # only QC samples includded
        self._bi_files = [x for x in glob.glob(path + '/*QC0*.mzXML')]          # only BI samples includded
        self._sizes = []
        
    def show():
        pass
    
    @property
    def ms_files(self):
        return self._ms_files
    
    @ms_files.setter
    def ms_files(filenames: list):
        self._ms_files = filenames

    @property
    def worklist(self):
        return self._worklist

    @worklist.setter
    def workist(self, worklist: pd.DataFrame):
        self._worklist = worklist
    
    def read_worklist(self, path_worklist):
#         worklist = pd.read_csv(filename)
        self._worklist = pd.read_csv(path_worklist, skiprows=1)
        isin = [self._worklist['File Name'][k] in self._filenames for k in range(len(self._worklist))]
        self._worklist = self._worklist[ isin ]
    

    def gui(self):
        pass
```
